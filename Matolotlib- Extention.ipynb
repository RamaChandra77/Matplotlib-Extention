### PIE CHART

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0.5, 0, 0) # explode 1st slice (i.e. 'A')

# Plot
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
        
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Exploded Pie Chart Example')
plt.show()
```

###### In a pie chart, the explode parameter allows you to "pull out" or "explode" one or more slices from the pie to emphasize them. This can be useful for highlighting specific data points.s.

### Explanation

##### 1. Data to plot:
###### .: Labels for each slice.
###### .sizes : The size of each slice.
###### .colors : Colors for each slice.
###### .explode : A tuple indicating which slice to pull out. In this case, the first slice ('A') is pulled out by 0.1.

##### 2. Plot:
###### .plt.pie() : Plots the pie chart.
###### .sizes : The data to plot.
###### .explode : Specifies the fraction to offset a slice. Here, (0.1, 0, 0, 0)
###### .means the first slice will be pulled out by 0.1.
###### .labels : The labels for each slice.
###### .colors : Colors for each slice.
###### .autopct='%1.1f%%' : Formats the value of each slice as a percentage.
###### .shadow=True : Adds a shadow to the pie chart.
###### .startangle=140 : Starts the pie chart at a specified angle.
###### .plt.axis('equal') : Ensures the pie is drawn as a circle.

##### 3. Display:
###### .plt.show() : Displays the plot.

#### Customizing the Explode Effect
###### You can customize which slices to explode and by how much. For example, to explode the second and fourth slices:
###### explode = (0, 0.1, 0, 0.2) # explode 2nd slice ('B') and 4th slice ('D')


### BOX PLOT

###### In a box plot, setting notch=True adds a notch around the median, providing a visual way to assess the significance of differences between medians. The notched box plot displays a confidence interval around the median, making it easier to compare the medians of different data sets.ts.

### Purpose of the Notch

###### The notch represents a confidence interval around the median. If the notches of two box plots do not overlap, it suggests that the medians of the two data sets are significantly different at the specified confidence level.



### Explanation

###### 1. notch=True : Adds a notch to the box plot.
###### 2. Data Generation: Three sets of normally distributed data are generated.
###### 3. Box Plot Creation: The ax.boxplot() function creates a box plot with notches.
###### 4. Visualization: The plot displays the median, quartiles, and the notches representing the confidence intervals around the medians.

### Interpretation

###### .Median: The line inside the box represents the median.
###### .Notch: The indentation around the median shows the confidence interval. If notches from different box plots do not overlap, it suggests significant differences between the medians.
###### .Box: Represents the interquartile range (IQR).
###### .Whiskers: Extend to the data points within 1.5 * IQR from the quartiles.
###### .Outliers: Data points beyond the whiskers are plotted as individual points.

### Use Case

###### Notched box plots are particularly useful when comparing the medians of multiple groups or data sets, providing a visual indication of whether the differences in medians are statistically significant.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(80, 30, 200)

# Create a figure
fig, ax = plt.subplots(figsize=(10, 6))

# Create a notched box plot
ax.boxplot([data1, data2, data3], notch=True, patch_artist=True)

# Set plot title and labels
ax.set_title('Notched Box Plot Example')
ax.set_xlabel('Data Set')
ax.set_ylabel('Value')

# Show the plot
plt.show()
```

###### Capping in a box plot refers to the representation of data points that are considered outliers.In a box plot, a "cap" usually refers to the whiskers, which are the lines extending from  the top and bottom of the box. These whiskers indicate the range of the data excluding outliers.



### Understanding Box Plots

###### A box plot, also known as a box-and-whisker plot, summarizes a data set by showing its minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It also identifies outliers.
###### 1. Median (Q2): The middle value of the data set.
###### 2. First Quartile (Q1): The median of the lower half of the data set.
###### 3. Third Quartile (Q3): The median of the upper half of the data set.
###### 4. Interquartile Range (IQR): The range between Q1 and Q3 (i.e., Q3 - Q1).

### Whiskers and Capping

###### Whiskers: The lines extending from the top and bottom of the box to the highest and lowest values within 1.5 * IQR from the Q1 and Q3, respectively.###### Whiskers: The lines extending from the top and bottom of the box to the highest and lowest values within 1.5 * IQR from the Q1 and Q3, respectively.points are considered outliers.



### Outliers

###### Outliers: Data points that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR.In some cases, whiskers are drawn to the maximum and minimum data points that are not considered outliers, while outliers are plotted as individual points beyond the whiskers.nts.s.



###### In this example:
###### .The box represents the interquartile range (IQR).
###### .The line inside the box represents the median (Q2).
###### .The "caps" (or whiskers) extend to the maximum and minimum values within 1.5 * IQR from the quartiles.
###### .Outliers beyond these caps are plotted as individual points.

### Summary

###### Capping in a box plot refers to the limits of the whiskers that represent the range of data,excluding outliers. These caps help in identifying the spread and distribution of the data as well as spotting outliers beyond the typical range.
